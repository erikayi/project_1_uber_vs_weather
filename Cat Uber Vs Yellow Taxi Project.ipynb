```python
r = requests.get(url)
yellow_x = r.json()
Yellow_Taxi_df = pd.DataFrame(yellow_x)
Yellow_Taxi_df.head()

Yellow_Taxi_df["pickup_datetime"]=pd.to_datetime(Yellow_Taxi_df.pickup_datetime)
Yellow_Taxi_df['new_date'] = [d.date() for d in Yellow_Taxi_df['pickup_datetime']]
Yellow_Taxi_df['new_time'] = [d.time() for d in Yellow_Taxi_df['pickup_datetime']]

Yellow_Taxi_df["Month"]= Yellow_Taxi_df.pickup_datetime.dt.month
Yellow_Taxi_df
```

```python
april=pd.read_csv(csv_path)
uber_april=pd.DataFrame(april)

may=pd.read_csv(csv_path1)
uber_may=pd.DataFrame(may)

june=pd.read_csv(csv_path2)
uber_june=pd.DataFrame(june)

july=pd.read_csv(csv_path3)
uber_july=pd.DataFrame(july)

aug=pd.read_csv(csv_path4)
uber_aug=pd.DataFrame(aug)

sept=pd.read_csv(csv_path5)
uber_sept=pd.DataFrame(sept)
```
